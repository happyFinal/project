```python
import pandas as pd
from glob import glob
import pickle
import re
```

```python
file_path = glob('*.gzip')[0]
file_path
```

```python
df = pd.read_parquet(file_path)
df
```

```python
condition = df["장르"] == "뮤지컬"
df.loc[condition, "가사"]
```

```python
df["장르"].value_counts()
```

# 전처리

```python
df["가사"] = df["가사"].map(lambda x : re.sub("[^a-zA-Zㄱ-ㅎ가-힣0-9\n ]", "", x).strip()) # 영어, 한글, 숫자, 개행문자, 공백만 남기고 제거
df["가사"] = df["가사"].map(lambda x : re.sub("\s{2,}", " ", x)) # 공백 2개 이상 있으면 1개로
df = df[df["가사"].map(lambda x : len(x) > 10)] # 위 두 개 전처리 후 10글자가 안되는 것 제거
df = df.reset_index(drop=True) # 인덱스 초기화
df.shape
```

# 샘플링

```python
df_ballad_1000 = df[df["장르"].str.contains("발라드")].sample(1000)
df_ballad_1000.head()
```

```python
df_ballad_1000["장르"].value_counts()
```

```python
df_rock_1000 = df[df["장르"].str.contains("록메탈")].sample(1000)
df_rock_1000.head()
```

```python
df_dance_1000 = df[df["장르"].str.contains("댄스")].sample(1000)
df_dance_1000.head()
```

```python
df_hiphop_1000 = df[df["장르"].str.contains("힙합")].sample(1000)
df_hiphop_1000.head()
```

```python
df_folk_1000 = df[df["장르"].str.contains("포크블루스")].sample(1000)
df_folk_1000.head()
```

```python
df_rnb_1000 = df[df["장르"].str.contains("RNB")].sample(1000)
df_rnb_1000.head()
```

```python
df_t_1000 = df[df["장르"].str.contains("트로트")].sample(1000)
df_t_1000.head()
```

```python
test_df = pd.concat([df_ballad_1000, df_dance_1000, df_rock_1000, df_hiphop_1000, df_rnb_1000, df_t_1000, df_folk_1000])
test_df.shape
```

```python
test_df["장르"].value_counts()
```

```python
test_df
```

```python
l = test_df["가사"].str.split("\n") 
l
```

```python
for i in l[:5]:
    print(i)
```

```python
lst = []
lyrics = test_df["가사"].str.split("\n")
for sentence in lyrics:
    for word in sentence:
        lst.append(word)
```

```python
# 중복 제거
lst = list(set(lst))
print(len(lst))
lst
```

```python
with open("test_data.txt", "w") as f:
    for element in lst:
        f.write(element + "\n")
    f.close()
```

```python
# pickle로 저장
with open("test_data.txt", 'wb') as lf:
    pickle.dump(lst, lf)
```
